```python
import pandas as pd
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df.head()
```

### SALARY PARSING

```python
df.info()
```

```python
# TO REMOVE SALARY ESTIMATE WHICH IS -1
df = df[df['Salary Estimate'] != '-1']
```

```python
df['Salary Estimate'][0].split('(')[0]
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary.head()
```

```python
salary_minus_kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
salary_minus_kd.head()
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
salary_digit_only = salary_minus_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
df['min_salary'] = salary_digit_only.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = salary_digit_only.apply(lambda x: int(x.split('-')[1]))
```

```python
df['avg_salary'] = (df.min_salary + df.max_salary) / 2
```

```python
df.head()
```

### COMPANY NAME TEXT ONLY

```python
df.head()
```

```python
df['Company Name'][0].split('\n')[0]
```

```python
df['company_name_txt'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
```

```python
df.head()
```

### STATE FIELD

```python
df.head()
```

```python
df['Location'][0].split(',')[1]
```

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df.head()
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0 , axis=1)
```

```python
df.head()
```

### AGE OF COMPANY

```python
df['age_of_company'] = df['Founded'].apply(lambda x: x if x<1 else 2020-x)
```

```python
df.head()
```

### PARSING OF JOB (PYTHON, SPARK, AWS, EXCEL ETC...)

```python
df['python_skill'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['spark_skill'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws_skill'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['excel_skill'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
df.head()
```

### SAVING DATA TO CSV

```python
df.columns
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```
